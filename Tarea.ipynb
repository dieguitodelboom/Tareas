```python
geocodes = {}
```

```python
def geocodificar(ciudad):
    texto = ciudad 
    x = geocodes.get(texto,"")
    if x == "":
        x = geocoder.komoot(texto).latlng
        geocodes[texto] = x
    return x 
```

```python
mymap = folium.Map()

for index, row in df.iterrows():
    ORIGEN_PROVINCIA = row.ORIGEN_PROVINCIA
    DESTINO_PROVINCA = row.DESTINO_PROVINCIA
    p1 = geocodificar(ORIGEN_PROVINCIA)
    p2 = geocodificar(DESTINO_PROVINCIA)
    points = [p1,p2]
    folium.PolyLine(points, color="green", wright=2.5).add_to(mymap)
    if index > 2000:
        break
mymap
```
